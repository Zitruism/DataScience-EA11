```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
import numpy as np
import pandas as pd

# Load dataset
df = pd.read_csv("https://raw.githubusercontent.com/Zitruism/DataScience-EA11/master/Dataset.csv", delimiter=";")
df




```

```python
# Replace string with numerical values since the function won't accept strings.
df = df.replace('<=30', 0)
df = df.replace('31…40', 1)
df = df.replace('>40', 2)
df = df.replace('No', 0)
df = df.replace('Yes', 1)
df = df.replace('Low', 0)
df = df.replace('Medium', 1)
df = df.replace('High', 2)
df = df.replace('Fair', 0)
df = df.replace('Excellent', 1)
df
```

```python
feature_cols = ['Age','Income','Student','Credit_rating']



X = df[feature_cols]
y = df.Buys_computer
 
tree_clf = DecisionTreeClassifier(max_depth=4)
tree_clf.fit(X, y)
```

```python
# Export graphdata into file
from sklearn.tree import export_graphviz
 
export_graphviz(
         tree_clf,
         out_file="tree.dot",
         feature_names=feature_cols,
         rounded=True,
         filled=True
 )
```
